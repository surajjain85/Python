# Write a Pandas program to join the two given dataframes along rows and assign all data.

```python
import pandas as pd

student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})
```

```python
print("Original DataFrames:")
print(student_data1)
print("-------------------------------------")
print(student_data2)
print("\nJoin the said two dataframes along rows:")
```

```python
result_data=pd.concat([student_data1,student_data2],axis=1)
result_data
```

# Write a Pandas program to join the two dataframes using the common column of both dataframes

```python
import pandas as pd
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})
```

```python
print("student data 1")
print(student_data1)
print("student data 2")
print(student_data2)
```

```python
merge_data= pd.merge(student_data1, student_data2,on ='student_id', how='inner')
merge_data
```

# Write a Pandas program to join the two dataframes with matching records from both sides where available.

```python
merge_data= pd.merge(student_data1, student_data2,on ='student_id', how='outer')
merge_data
```

# Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.

```python
import pandas as pd
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})
```

```python
print("data1")
print(data1)
print("data2")
print(data2)
```

```python
merge_data1= pd.merge(data1,data2, on=('key1','key2'), how= 'right')
merge_data1
```
