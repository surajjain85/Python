# Categorical, Dummy Variables, One Hot Encoding

1. using pandas get_dummies
2. Using sklearn OneHotEncod

```python
import pandas as pd
```

```python
df=pd.read_csv("D:\\SimpliLearn\\Python\\ML\\Dummy_onehotencoding.csv")
df
```

```python
#as of now we have the nameof town but we need to numeric data like 1,2,3 but this will not work
#here we need oneHotEncoding concet, which will create new colums of towns as dummy variables (like in this delhi,gurgaon,noida) as 0,1

#this will create delhi as 1,gurgaon 0 and noida 0 in column 1
#use dummy variable function to convert the categorical variables into numeric
dummies=pd.get_dummies(df.town)
```

```python
#now here i'll merge the two dataset df and dummies
merged= pd.concat([df,dummies],axis='columns')
merged
```

```python
#now the steps to drop the town column as we have dummy variable columns
#next i'll drop the any one dummy variable, here iam dropping delhi as delhi is 1 and other 2 are 0,0, so if gurgaon
#and noida are 0 then it automatically execute delhi

final= merged.drop(['town','delhi'],axis='columns')
final

#although sklearn model automatically drops one dummy variabl column but its a good practice that u drop it yourself (dummyvariable_col -1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#now its time to give x(independent) and y(dependent) variable
#x is other than price, so here i'll drop price
#y is price

x= final.drop('price',axis='columns')
x
```

```python
y=final.price
y
```

```python
#time to train the model
model.fit(x,y)
```

```python
#prediction starts
#input for prediction would be x, (area,gurgaon,noida)
#here iam predicting price for gurgaon with 2600,1,0
model.predict([[2600,1,0]])


```

```python

#now for noida
model.predict([[2800,0,1]])
```

```python
#for delhi
model.predict([[3400,0,0]])
```

```python
#check theaccuracy of our mmodel use score method
model.score(x,y)
```
