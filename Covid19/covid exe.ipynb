Get the latest number of confirmed, deaths, recovered and active cases of Novel Coronavirus Country wise

```python
covid_data= pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-17-2020.csv')
```

```python
covid_data['Active Cases']=covid_data['Confirmed']-covid_data['Deaths']-covid_data['Recovered']
covid_data['Active Cases']
```

```python
result=covid_data.groupby('Country/Region')['Confirmed','Deaths','Recovered','Active Cases'].sum().reset_index()
result
```
